```python
data = None

for side in sides:
    if not data is None:
        data =pd.concat((data, pd.read_csv(os.path.join ( BASE_PATH,side),skiprows =10)))

    else:
        data= pd.read_csv(os.path.join ( BASE_PATH,side),skiprows =10)
```
